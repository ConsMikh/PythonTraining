```python
# Объединение слева таблиц REQ, AGR по полю 'AGR_GUID', потом слева с ADR по полю 'REQ_GUID'
req_agr = pd.merge(hist_req, hist_agr, on = 'AGR_GUID', how = 'left')
req_agr_adr = pd.merge(req_agr, hist_addr, on = 'REQ_GUID', how = 'left')
req_agr_adr.info()
```

```python
# Для реализации фильтра нужны поля "ARG.Тип" и "AGR.Дата подписания/аванса", которых нет выборке
# Вместо поля "AGR.Дата подписания/аванса" взято совпадающее поле 'ДатаПодписанияДоговораЗаявителем'

req_agr_adr = req_agr_adr[['Дата_x','SETTL_FID_x',
                            'REQ_GUID',
                            'ДатаПускаГаза',
                            'ДатаСтроительстваГазопроводВвода',
                            'ДатаОкончанияДействия', 
                            'ДатаПодписанияДоговораЗаявителем', 
                            'СтоимостьКЧ_ПИР', 
                            'СтоимостьКЧ_СМР', 
                            'СтоимостьКЧ_ТМЦ', 
                            'СтоимостьТО', 
                            'СтоимостьВДГО', 
                            'СтоимостьМонтажБезГаза', 
                            'ДатаАванса']]
```

```python
# Преобразование дат к типу Datetime
req_agr_adr['Дата_x'] = pd.to_datetime(req_agr_adr['Дата_x'])
req_agr_adr['ДатаПускаГаза'] = pd.to_datetime(req_agr_adr['ДатаПускаГаза'])
req_agr_adr['ДатаСтроительстваГазопроводВвода'] = pd.to_datetime(req_agr_adr['ДатаСтроительстваГазопроводВвода'])
req_agr_adr['ДатаОкончанияДействия'] = pd.to_datetime(req_agr_adr['ДатаОкончанияДействия'])
req_agr_adr['ДатаПодписанияДоговораЗаявителем'] = pd.to_datetime(req_agr_adr['ДатаПодписанияДоговораЗаявителем'])
req_agr_adr['ДатаАванса'] = pd.to_datetime(req_agr_adr['ДатаАванса'])
```

```python
req_agr_adr['Дата подписания/аванса'] = req_agr_adr['ДатаПодписанияДоговораЗаявителем']
index = req_agr_adr['Дата подписания/аванса'].loc[req_agr_adr['Дата подписания/аванса'].isna()].index
req_agr_adr['Дата подписания/аванса'].loc[index] = req_agr_adr['ДатаАванса'].loc[index]
```

```python
# Создание поля Тип.

req_agr_adr['AGR.Тип'] = req_agr_adr['СтоимостьКЧ_ПИР']+req_agr_adr['СтоимостьКЧ_СМР']+req_agr_adr['СтоимостьКЧ_ТМЦ']+req_agr_adr['СтоимостьТО']+req_agr_adr['СтоимостьВДГО']+req_agr_adr['СтоимостьМонтажБезГаза']
req_agr_adr['Тип'] = 'ТП0'
req_agr_adr['Тип'].loc[req_agr_adr['AGR.Тип']>0] = 'КД'
req_agr_adr['Тип'].value_counts()
```

```python
# Функция для посчета просроченных договоров втупую. Она медленная, но понятная. Ориентировался на ее результаты, 
# когда реализовывал 4-х этажное условие средставами pandas

def get_req_err(i):
    fpd = 0
    DateNow = pd.to_datetime('today').normalize()
    DateFinAct = DateNow if pd.isna(req_agr_adr['ДатаОкончанияДействия'].iloc[i]) else req_agr_adr['ДатаОкончанияДействия'].iloc[i]
    DateAvans = req_agr_adr['Дата подписания/аванса'].iloc[i]
    DateGazIn = req_agr_adr['ДатаСтроительстваГазопроводВвода'].iloc[i]
    DatePusk = req_agr_adr['ДатаПускаГаза'].iloc[i]
    Type = req_agr_adr['Тип'].iloc[i]
    if pd.isna(DateAvans): fpd =0
    elif Type == 'ТП0':
        if (DateFinAct < DateNow) & (pd.isna(DateGazIn)): fpd = 1
        else: fpd =0
    elif Type == 'КД':
        if (DateFinAct < DateNow) & (pd.isna(DatePusk)): fpd = 1
        else: fpd = 0
    return fpd
```

```python
req_agr_adr['st_slow'] = req_agr_adr.index
req_agr_adr['st_slow'] = req_agr_adr['st_slow'].apply(get_req_err)
req_agr_adr['st_slow'].value_counts()
```

```python
# Рассчет просроченных договоров функциями pandas
DateNow = pd.to_datetime('today').normalize()

req_agr_adr['st_fast'] = 0
req_agr_adr['DateFinAct'] = req_agr_adr['ДатаОкончанияДействия']
req_agr_adr['DateFinAct'].apply(lambda x: DateNow if pd.isna(x) else x)

req_agr_adr['st_fast'].loc[(req_agr_adr['Дата подписания/аванса'].notna())&(req_agr_adr['Тип']=='ТП0')&(req_agr_adr['DateFinAct'] < DateNow) & (req_agr_adr['ДатаСтроительстваГазопроводВвода'].isna())] = 1
req_agr_adr['st_fast'].loc[(req_agr_adr['Дата подписания/аванса'].notna())&(req_agr_adr['Тип']=='КД')&(req_agr_adr['DateFinAct'] < DateNow) & (req_agr_adr['ДатаПускаГаза'].isna())] = 1
```

```python
req_agr_adr['st_fast'].value_counts()
```

```python
# Формирование итогового списка для фильтра 'Договор просрочен' 
req_agr_adr_dog_list = req_agr_adr[['Дата_x','SETTL_FID_x','REQ_GUID']].loc[req_agr_adr['st_fast']==1]
req_agr_adr_dog_list['Error_type'] = 'Договор просрочен'
req_agr_adr_dog_list['Error_count'] = 1
req_agr_adr_dog_list.info()
```
