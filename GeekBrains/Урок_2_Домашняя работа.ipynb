```python
import pandas as pd
from functools import reduce
```

```python
users = pd.read_csv('Урок_2_Материалы/users.csv')
```

```python
users.head()
```

```python
subscribers = pd.read_csv('Урок_2_Материалы/subscribers.csv')
```

```python
m_camp = pd.read_csv('Урок_2_Материалы/marketing_campaign.csv')
```

```python
subscribers.head()
```

```python
m_camp.head()
```

```python
datasets = [m_camp, users, subscribers]
fulldata = reduce(lambda left,right: pd.merge(left, right, on = 'user_id', how = 'left'), datasets)
```

```python
fulldata.head()
```

```python
fulldata.describe()
```

```python
fulldata.info()
```

```python
fulldata['date_served'] = pd.to_datetime(fulldata['date_served'])
```

```python
# Получение списка дат месяца из поля 'date_served'
# sorted(fulldata['date_served'].dt.day.dropna().unique().tolist())
# Берется поле, в нем через dt.day выбираются все числа всех месяцев
# dropna() - откидывает значения nan
# unique() - выбирает только уникальные значения
# tolist() - преобразует массив в спискок
# sorted() - упорядочивает по возрастанию
# Получается массив float
# Для преобразования этого массива к int применяются два способа
# 1. Циклом по всем значениям с применением функции int
# 2. Функцией map. Она дает объект mapobject, который преобразуется к списку

lst = [int(x) for x in sorted(fulldata['date_served'].dt.day.dropna().unique().tolist())]
print (lst)

lst1 = list(map(int, sorted(fulldata['date_served'].dt.day.dropna().unique().tolist())))
print (lst1)
```
